```python
1. Исследовать на условный экстремум функцию
U=3−8x+6y, 
если
x2+y2=36

L(x, y, λ) = 3 - 8x + 6y + λ * (x^2 +y^2 = 36) 

L'(x) = -8 + λ * 2x = 0
L'(y) = 6 + λ*2y = 0
L'(λ) = x^2 + y^2 - 36 = 0

-8 + λ * 2x = 0
x = 8/2λ

6 + λ*2y = 0
y = -6/2λ

(8/2λ)^2 + (-6/2λ)^2 =36

x = 4λ
y = -3λ
100/4λ^2 = 36
λ^2 = 25/9
λ = +- 5/3

+ (20/3, -5, 5/3)
- (-20/3, 5, -5/3)

L''(x) = 2λ
L''(y) = 2λ
L''(λ) = 0

L''(xy) = L''(yx) = 0
L''(xλ) = L''(λx) = 2x
L''(yλ) = L''(xλ) = 2y

0   2x  2y
2x  2λ  0
2y  0   2λ

    2λ  0         2x  0         2x  2λ
0 * 0   2λ - 2x * 2y  2λ + 2y * 2y  0    
= (-2x) * 4xλ + 2y * (-4yλ) 
= -8λ(x^2 + y^2) 

x2+y2=36

Δ = -8λ * 36 = -288

λ = 5/3, Δ < 0, => (20/3, -5, 5/3), min
λ = -5/3, Δ > 0, => (20/3, -5, 5/3), max
```

```python
2. Исследовать на условный экстремум функцию

U=2x^2+12xy+32y^2+15, 
если

x^2+16y^2=64

L(x, y, λ) = 2x^2 + 12xy + 32y^2 + 15 + λ * (x^2 + 16y^2 = 64)

L'(x) = 4x + 12y + λ * 2x = 0
L'(y) = 12x + 64y + λ * 16y = 0
L'(λ) = x^2 + 16y^2 - 64 = 0

x^2 + 16y^2 = 64
x + 4y = 8
x = 8 - 4y
x = 8 - 4*(1/λ)
x = 8 - 4/λ
x = 4

12*(8 - 4y) + 64y + λ * 16y = 0
96 - 48y + 64y + 16yλ = 0
8 - 3y + 4y +yλ = 0
8 - 8yλ = 0
-yλ = -1
y = 1/λ

(4)^2 + 16(1/λ)^2 - 64 = 0
4 + 4/λ - 8 = 0
4/λ = 4
λ = +-1

+ (4, 1, 1)
- (-4, -1, -1)

L''(x) = 4 + 2λ = 0
L''(y) = 64 + 16λ = 0
L''(λ) = x^2 + 16y^2 - 64 = 0

x = λ/2
y = λ/4
λ = 0

L''(x) = 2λ
L''(y) = 2λ
L''(λ) = 0

L''(xy) = L''(yx) = 0
L''(xλ) = L''(λx) = 2x
L''(yλ) = L''(xλ) = 2y

0   2x  2y
2x  2λ  0
2y  0   2λ

    2λ  0         2x  0         2x  2λ
0 * 0   2λ - 2x * 2y  2λ + 2y * 2y  0    
= (-2x) * 4xλ + 2y * (-4yλ) 
= -8λ(x + y)^2 

x^2+16y^2=64
x + y = 4

Δ = -8λ * 8 = -64λ

λ = 1, Δ < 0, => (4, 1, 1), min
λ = -1, Δ > 0, => (-4, -1, -1), max
```
