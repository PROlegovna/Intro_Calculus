```python
## з∗)limx→+∞ln(x2−x+1)ln(x10+x+1) 

from sympy import limit, factorial, oo, log
from sympy.abc import x
expr = log((x ** 2) - x + 1) / log((x ** 10) + x +1)
limit_expr = limit(expr, x, oo)  
print (limit_expr)
```

```python
## 2*. На языке Python предложить алгоритм вычисляющий численно предел последовательности
## limn→+∞nn!−−√n

import numpy as np
factorial = np.uint64(1)
i = np.uint64()
for i in range(1,400):
    factorial *= i
    a = np.float64(i / (factorial ** (1 / i)))
    a = round (a, 10)
    print(f' Без упрощения i = {i}, {a}')

## Стремится к 2,7 
```
